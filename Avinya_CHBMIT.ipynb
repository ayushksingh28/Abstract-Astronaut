```python
n = len(df)

n_val = int(n * 0.02)
n_test = int(n * 0.02)
n_train = n - n_val - n_test
```

```python
n
```

```python
n_val, n_test, n_train
```

```python
df_train = df.iloc[:n_train]
df_val = df.iloc[n_train:n_train+n_val]
df_test = df.iloc[n_train+n_val:]
```

```python
idx = np.arange(n)
```

```python
np.random.seed(2)
np.random.shuffle(idx)
```

```python
df_train = df.iloc[idx[:n_train]]
df_val = df.iloc[idx[n_train:n_train+n_val]]
df_test = df.iloc[idx[n_train+n_val:]]
```

```python
count = [len(df_train), len(df_val), len(df_test)]
categories = ['df_train', 'df_val', 'df_test']

plt.bar(categories, count, color=['blue', 'green', 'red'])

plt.xlabel('Datasets')
plt.ylabel('Count')
plt.title('Distribution of Data count')

plt.show()
```

```python
df_train = df_train.reset_index(drop=True)
df_val = df_val.reset_index(drop=True)
df_test = df_test.reset_index(drop=True)
```

```python
X_train = df_train.drop('Outcome', axis=1)
y_train = df_train['Outcome']
X_train_norm = pd.DataFrame(normalize(X_train), columns=X_train.columns)

# Combine the normalized features with the target column
df_train = pd.concat([X_train_norm, y_train], axis=1)
```

```python
X_test = df_test.drop('Outcome', axis=1)
y_test = df_test['Outcome']
X_test_norm = pd.DataFrame(normalize(X_test), columns=X_test.columns)

# Combine the normalized features with the target column
df_test = pd.concat([X_test_norm, y_test], axis=1)
```

```python
X_val = df_val.drop('Outcome', axis=1)
y_val = df_val['Outcome']
X_val_norm = pd.DataFrame(normalize(X_val), columns=X_val.columns)

# Combine the normalized features with the target column
df_val = pd.concat([X_val_norm, y_val], axis=1)
```

```python
from sklearn.model_selection import KFold
model = RandomForestClassifier()
n_estimators = [100, 250]
max_features = ['sqrt']
# define grid search
grid = dict(n_estimators=n_estimators,max_features=max_features)
cv = KFold(n_splits=10, random_state=None)
grid_search = GridSearchCV(estimator=model, param_grid=grid, n_jobs=-1, cv=cv, scoring='accuracy',error_score=0)
grid_result = grid_search.fit(X_train, y_train)
# summarize results
print("Best: %f using %s" % (grid_result.best_score_, grid_result.best_params_))
means = grid_result.cv_results_['mean_test_score']
stds = grid_result.cv_results_['std_test_score']
params = grid_result.cv_results_['params']
for mean, stdev, param in zip(means, stds, params):
    print("%f (%f) with: %r" % (mean, stdev, param))
```

```python
model=model.fit(X_train,y_train)
```

```python
pred=model.predict(X_val)
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
print("Confusion Matrix:\n",confusion_matrix(y_val,pred))
print("Classification Report:\n",classification_report(y_val,pred))
```
